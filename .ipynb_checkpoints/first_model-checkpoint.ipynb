```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
%matplotlib inline
import os
```

```python
os.getcwd()
```

```python
# Load data

dataset = pd.read_csv("../toobig/cc_fraud_data/data.csv")
```

```python
# Check for Null values
dataset.isnull().sum()
```

```python
null_data = dataset[dataset.isnull().any(axis=1)]
null_data
```

```python
dataset = dataset.dropna()
```

```python
g = sns.distplot(dataset["Time"], color="m", label="Skewness : %.2f"%(dataset["Time"].skew()))
g = g.legend(loc="best")
```

```python
plt.figure(figsize = (20,15))
g = sns.heatmap(dataset.corr(), annot=True, fmt = ".2f", cmap = "coolwarm")
```

```python
g = sns.FacetGrid(dataset, col='Class')
g = g.map(sns.distplot, "Time")
```
